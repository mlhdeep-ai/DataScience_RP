```python
dia = pd.read_csv('Diabetes.csv')
```

```python
dia
```

```python
dia = dia.rename(columns = {' Number of times pregnant' : 'num_times_preg'})
```

```python
dia
```

```python
dia = dia.rename(columns = {' Plasma glucose concentration' : 'plasma_gluc_conct', ' Diastolic blood pressure' : 'diast_blood_pres', ' Triceps skin fold thickness' : 'tric_skin_fold_thick', ' 2-Hour serum insulin' : '2h_serum_insul', ' Body mass index' : 'bmi',
                            ' Diabetes pedigree function' : 'diab_pedig_func', ' Age (years)' : 'age', ' Class variable' : 'output'})
```

```python
dia
```

```python
dia['output'] = dia['output'].str.replace('YES', '1')
dia['output'] = dia['output'].str.replace('NO', '0')
```

```python
dia
```

```python
corr = dia.corr()
```

```python
corr
```

```python
sn.heatmap(corr, annot = True)
```
